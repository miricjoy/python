```python
#5.Write a Python program to check if a number is prime. Expected Output (for input 7):
#7 is a prime number

num=int(input("Enter a number"))


if num==0 or num==1:
    print(f"{num} is a prime number")
elif num >1 :
    for i in range( 2,num):
        if num%i == 0:
            print(f"{num} is not a prime number")
            break
        
    print(f"{num} is a prime number")
        



```
