```python
def OddEven():
    num= int(input("enter a number"))
    if (num%2==0):
        print("even")
        answer="divisible  by 2"
    else:
        print("odd")
        answer="not divisible by 2"
    return answer
```

```python

def eligibility():

    gender=input("enter your gender")
    age=int(input("enter your age"))
    
    if (gender=="Male" and age>=21):
            print("eligible for marriage")
    elif (gender=="Female" and age>=18):
            print("eligible for marriage")
    else:
            print("not eligible")
```

```python
eligibility()
```

```python
def percentageCalculator():
    subject1=int(input("Enter subject1 marks"))
    subject2=int(input("Enter subject2 marks"))
    subject3=int(input("Enter subject3 marks"))
    subject4=int(input("Enter subject4 marks"))
    subject5=int(input("Enter subject5 marks"))
    print("Subject1 =",subject1 , "\n Subject2 =",subject2 , "\n Subject3 =",subject3 , "\n Subject4 =",subject4 , "\n Subject5 =",subject5)
    total= subject1+subject2+subject3+subject4+subject5
    percentage= total/5
    print("Total =", total, '\n Percentage =' ,percentage)
```

```python
percentageCalculator()
```

```python
def triangle():
    height=int(input("Enter the height"))
    breadth=int(input("Enter the breadth"))
    side1=int(input("Enter the length of side1"))
    side2=int(input("Enter the length of side2"))
    side3=int(input("Enter the length of side3"))
    area = (height*breadth)/2
    perimeter=side1+side2+side3
    print("area:",area)
    print("perimeter:",perimeter)
```

```python
triangle()
```
